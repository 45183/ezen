```python
import tensorflow as tf
from tensorflow import keras
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# MNIST dataset download 손글씨 데이터
mnist = keras.datasets.mnist

(x_train, y_train), (x_test, y_test) = mnist.load_data()
x_train, x_test = x_train / 255.0, x_test / 255.0
```

```python
# 데이터 확인

import random
import matplotlib.pyplot as plt

samples = random.choices(population=range(0, len(y_train)), k=25)

plt.figure(figsize=(14, 12))
for idx, n in enumerate(samples):
    plt.subplot(5, 5, idx+1)
    plt.imshow(x_train[n].reshape(28, 28), cmap='Greys', interpolation='nearest')
    plt.title('Label : ' + str(y_train[n]))
    plt.axis('off')
plt.show()
```

```python
# Model 생성, compile
model = tf.keras.models.Sequential([
  tf.keras.layers.Flatten(input_shape=(28, 28)),
  tf.keras.layers.Dense(128, activation='relu'),
  tf.keras.layers.Dropout(0.2),
  tf.keras.layers.Dense(10, activation='softmax')
])

#모델 컴파일 코드

model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

(Model 생성, compile) 코드는 TensorFlow 라이브러리를 사용하여 Sequential 모델을 만드는 것입니다. 모델은 다음과 같은 레이어로 구성됩니다:

- Flatten 레이어 : 이 레이어는 입력 데이터를 1차원 형태로 평탄화합니다. 예제에서는 28x28 크기의 이미지 데이터를 입력으로 사용하므로, 이 레이어는 입력 데이터를 784 (28x28) 크기의 1차원 배열로 변환합니다.
- Dense 레이어 : 이 레이어는 완전 연결층(fully-connected layer)이며, 입력 데이터와 출력 데이터를 모두 연결합니다. 첫 번째 Dense 레이어는 128개의 노드로 구성되어 있으며, relu 활성화 함수를 사용합니다. 이 레이어는 입력 데이터의 특성(feature)을 학습합니다.
- Dropout 레이어 : 이 레이어는 입력 데이터의 일부를 무작위로 제거하여 과적합(overfitting)을 방지합니다. 이 예제에서는 Dropout 비율을  0.2로 설정했으므로, 입력 데이터의 20%가 제거됩니다.
- 두 번째 Dense 레이어 : 이 레이어는 출력층(output layer)이며, 10개의 노드로 구성되어 있습니다. softmax 활성화 함수를 사용하여 각 클래스(0~9)에 대한 확률값을 계산합니다. 즉, 이 레이어는 입력 데이터가 어떤 클래스에 속할 확률을 출력합니다. 따라서, 128과 10은 각각 첫 번째 Dense 레이어와 두 번째 Dense 레이어의 노드 수(node)를 나타냅니다. 첫 번째 Dense 레이어는 128개의 노드를 가지며, 두 번째 Dense 레이어는 10개의 노드를 가지므로, 출력층은 10개의 클래스(0 - 9)를 구분할 수 있습니다

- 만약 128 dense를 512 dense로 변경한다면, 모델의 첫 번째 Dense 레이어의 노드 수가 4배 증가하게 됩니다. 이 경우 모델이 학습하는 매개변수(parameter)의 수도 증가하게 됩니다. 이로 인해 모델의 복잡도가 증가하고, 더 많은 계산 리소스와 더 많은 데이터를 필요로 할 수 있습니다.
- 더 많은 노드를 가진 Dense 레이어를 사용하면, 모델은 더 많은 특성을 학습할 수 있으며, 훈련 데이터에 더 잘 적합될 수 있습니다. 그러나 더 큰 모델은 과적합의 위험성이 높아질 수 있으므로, 충분한 데이터를 갖추지 못하면 성능이 저하될 수 있습니다. 따라서, 노드 수를 변경할 때는 모델의 복잡도와 데이터의 양을 고려하여 적절한 값을 선택해야 합니다.


(모델 컴파일) 코드는 TensorFlow 라이브러리를 사용하여 모델을 컴파일하는 것입니다. 모델을 컴파일하려면 최적화 알고리즘(optimizer), 손실 함수(loss function), 평가 지표(metrics)를 지정해야 합니다.

- optimizer='adam': 이 코드는 Adam 최적화 알고리즘을 사용하도록 설정합니다. Adam은 경사 하강법(Gradient Descent) 기반의 최적화 알고리즘 중 하나로, 학습률(learning rate)을 자동으로 조절하여 빠르고 안정적인 학습을 가능하게 합니다.
- loss='sparse_categorical_crossentropy': 이 코드는 희소 레이블(sparse label)을 가진 다중 클래스 분류(multi-class classification) 문제를 해결하기 위해, 희소 범주형 크로스 엔트로피(sparse categorical crossentropy) 손실 함수를 사용하도록 설정합니다. 손실 함수는 모델의 예측값과 실제 레이블 사이의 차이를 계산합니다.
- metrics=['accuracy']: 이 코드는 모델의 평가 지표로 정확도(accuracy)를 사용하도록 설정합니다. 정확도는 모델이 예측한 값 중 올바른 예측의 비율을 나타내는 지표입니다. 즉, 모델이 올바른 클래스를 얼마나 많이 예측했는지를 측정합니다. 따라서, 이 코드는 Adam 최적화 알고리즘, 희소 범주형 크로스 엔트로피 손실 함수, 그리고 정확도 평가 지표를 사용하여 모델을 컴파일합니다.

```python
# Training / Evaluation
result = model.fit(x_train, y_train, validation_data=(x_test, y_test), epochs=2)

model.evaluate(x_test,  y_test)
```

```python
import matplotlib.pyplot as plt

plot_target = ['loss', 'val_loss', 'accuracy', 'val_accuracy']

plt.figure(figsize=(12, 8))
for each in plot_target:
    plt.plot(result.history[each], label=each)
plt.legend()
plt.grid()
plt.show()
```

```python
score = model.evaluate(x_test, y_test)
print('Test loss : {}'.format(score[0]))
print('Test accuracy : {}'.format(score[1]))
```

```python
import os
from PIL import Image

# 우리 손글씨 파일 경로 설정
cur_dir = os.getcwd()
img_path = os.path.join(cur_dir, 'image.png')

# 파일 읽기
cur_img = Image.open(img_path)

# 28x28로 resize
cur_img = cur_img.resize((28, 28))
image = np.asarray(cur_img)


# color image일 경우 RGB 평균값으로 gray scale로 변경
try:
  image = np.mean(image, axis=2)
except:
  pass


# upload한 image는 흰 배경에 검은 글씨로 되어 있으므로, MNIST data와 같이 검은 배경에 흰 글씨로 변경
image = np.abs(255-image)


# 이 코드는 이미지 데이터를 0에서 255까지의 값에서 0에서 1까지의 값으로 정규화하는 작업을 수행합니다.
# astype 메소드를 사용하여 이미지 데이터의 데이터 타입을 float32로 변환한 뒤, 255로 나누어서 모든 값을 0과 1 사이의 값으로 조정합니다. 이 과정을 통해 모델이 학습하거나 예측할 때, 이미지 데이터의 값이 크게 차이나는 것을 방지하여 학습의 안정성을 높이고, 모델의 성능을 향상시킵니다.
# 이 코드는 이미지 데이터를 전처리하는 과정 중 하나이며, 딥러닝 모델에서 이미지 데이터를 다룰 때 매우 일반적인 작업 중 하나입니다.
# data preprocessing(255로 나눠줌)
image = image.astype(np.float32)/255.


# 출력
plt.imshow(image, cmap='gray')
plt.show()
```
