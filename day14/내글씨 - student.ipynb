```python
import tensorflow as tf
from tensorflow import keras
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# MNIST dataset download 손글씨 데이터
mnist = keras.datasets.mnist

(x_train, y_train), (x_test, y_test) = mnist.load_data()
x_train, x_test = x_train/255.0, x_test/255.0
```

```python
# 테스트 데이터 무작위로 불러서 데이터 확인

import random
import matplotlib.pyplot as plt
# 60000장의 사진 중 
samples = random.choices(population=range(0, len(y_train)), k=16) # 16개


# 지난 시간에 reshape가 딥러닝에 왜 중요하다고 했죠?
plt.figure(figsize=(14, 12)) # 개별 사진의 사이즈
for idx, n in enumerate(samples): # 개별 라벨과 이미지를 for문으로 소환
    plt.subplot(4, 4, idx+1) # 사진 4*4장 -> 16장 보여줘
    plt.imshow(x_train[n].reshape(28, 28), cmap='Greys', interpolation='nearest')
    plt.title('Label : ' + str(y_train[n]))
    plt.axis('off')
plt.show()
```

```python
# Model 생성, compile
# 앞에 dense랑 뒤에 dense 의미 무엇?
# model comiple 뜻 무엇?

model = tf.keras.models.Sequential([
  tf.keras.layers.Flatten(input_shape=(28, 28)),
  tf.keras.layers.Dense(128, activation='relu'),
  tf.keras.layers.Dropout(0.2),
  tf.keras.layers.Dense(10, activation='softmax')
])

model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

```python
# Training 뭐해야하죠?
result = model.fit(x_train, y_train, validation_data=(x_test, y_test), epochs=2) # epochs - 시도횟수

#평가는 어떻게 할까요?
model.evaluate(x_test,  y_test)
```
